```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# データの作成
x = np.arange(0, 6, 0.1) # 0 から6 まで0.1 刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)
# グラフの描画
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos") # 破線で描画
plt.xlabel("x") # x 軸のラベル
plt.ylabel("y") # y 軸のラベル
plt.title('sin & cos') # タイトル
plt.legend()
plt.show()
```

# パーセプトロン

パーセプトロンは、ローゼンブラッドというアメリカの研究者によって1957年に考案されたアルゴリズム。
また「アルゴリズム」という言葉がでてきたお。

アルゴリズムの実体はなんだ？ということで、Wikipedia先生に聞いてみると


- 数学、コンピューティング、言語学、あるいは関連する分野において、問題を解くための手順を定式化した形で表現したものを言う。
- 算法と訳されることもある。「問題」はその「解」を持っているが、アルゴリズムは正しくその解を得るための具体的手順および根拠を与える。さらに多くの場合において効率性が重要となる。
- コンピュータにアルゴリズムをソフトウェア的に実装するものがコンピュータプログラムである。人間より速く大量に計算ができるのがコンピュータの強みであるが、その計算が正しく効率的であるためには、正しく効率的なアルゴリズムに基づいたものでなければならない。

なるほど、問題を解くための具体的手順と根拠を与えるものか。アルゴリズムは根拠という意味を含むことはあまり意識したことがないな。

## パーセプトロンとは

パーセプトロンは、複数の信号を入力として受け取り、ひとつの信号を出力する。
2つの信号を入力として受け取るパーセプトロンの例を以下に示す。

![パーセプトロン](Perceptron.png)

$x_1$、$x_2$は入力信号、$y$は出力信号、$w_1$、$w_2$は重みを表し、図の○はニューロンやノードと呼ばれる。

入力信号は、ニューロンに送られる際にそれぞれに固有の重みが乗算され、ニューロンでは送られてきた信号の総和が計算され、その総和がある閾値を超えた場合にのみ$1$を出力します。これを

### ニューロンが発火する

と表現することもあります。

重みは、信号の流れやすさをコントロールする電気回路における抵抗と同様の働きをするお。

$$
\begin{equation}
y = \begin{cases}
0 & (w_{1}x_{1}+w_{2}x_{2}\ge\theta) \\
1 & (w_{1}x_{1}+w_{2}x_{2}<\theta)
\end{cases}
\end{equation}
$$
